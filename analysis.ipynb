Since the columns in the dataset are a total of 104, it would be overwhelming to perform analysis on all of them at once. Instead, It would be better to combine some features that hold the same meaning. For example, - lwb, ldm, cdm, rdm, rwb, lb, lcb, cb, rcb, rb are 9 defending position columns so, we can combine all these to one column `defensive_positions`. This would reduce the number of columns from 104 to 95.

- ls, st, rs, lw, lf, cf, rf, rw, lam, cam, ram, lm, lcm, cm, rcm, rm : `offensive_positions`

````python
def convert_int(x):
    try:
        if "+" in x:
            # Split the string into two parts and convert each part to an integer
            num1, num2 = map(int, x.split("+"))
            # Add the two numbers together and return the result
            return num1 + num2
        else:
            # Convert the string to an integer and return it
            return int(x)
    except (ValueError, TypeError):
        # Ignore the error and return the original string
        return x
````

````python
df['lwb'] = (df['lwb'].apply(convert_int)).fillna(0).astype(int)
df['ldm'] = (df['ldm'].apply(convert_int)).fillna(0).astype(int)
df['rdm'] = (df['rdm'].apply(convert_int)).fillna(0).astype(int)
df['cdm'] = (df['cdm'].apply(convert_int)).fillna(0).astype(int)
df['rwb'] = (df['rwb'].apply(convert_int)).fillna(0).astype(int)
df['lb'] = (df['lb'].apply(convert_int)).fillna(0).astype(int)
df['lcb'] = (df['lcb'].apply(convert_int)).fillna(0).astype(int)
df['cb'] = (df['cb'].apply(convert_int)).fillna(0).astype(int)
df['rcb'] = (df['rcb'].apply(convert_int)).fillna(0).astype(int)
df['rb'] = (df['rb'].apply(convert_int)).fillna(0).astype(int)
````

````python
# lwb, ldm, cdm, rdm, rwb, lb, lcb, cb, rcb, rb 
defensive_positions = (df['lwb'] + df['ldm'] + df['cdm'] + df['rdm'] + df['rwb'] + df['lb'] + df['lcb'] + df['cb'] + df['rcb'] + df['rb'])/10
defensive_positions
````

````python
# Displaying Object type value counts of dataset Columns including NANs

def val_count(df):
    
    num_to_display = 10 
    for colname in df.dtypes[df.dtypes == 'object'].index:
        val_counts = df[colname].value_counts(dropna=False) # to see NANs as well
        print(val_counts[:num_to_display])
        if (len(val_counts)>num_to_display):
            print(f' \n only displaying first {num_to_display} of {len(val_counts)} values')
        print('\n\n') # just for more spaces inbetween    
````

#### Q2 - How has the average player rating changed over time?

````python
df['dob'] = pd.to_datetime(df['dob'])
````

````python
df['year'] = df['dob'].apply(lambda x: x.year)
grouped_rating = df.groupby('year')['overall'].mean().reset_index()
grouped_rating.rename(columns={'overall':'average_rating'}, inplace=True)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16,6))
sns.lineplot(x='year', y='average_rating', data=grouped_rating)
````

#### Q4 - What is the distribution of player ratings by position?

```python

````python
grouped_position = df.groupby('player_positions')['overall'].mean().reset_index()
grouped_position.rename(columns={'overall':'avg_player_ratings'}, inplace=True) 
grouped_position.sort_values(by='avg_player_ratings', ascending=False, inplace=True)
grouped_position.head(15)

plt.figure(figsize=(16,8))
ax = sns.barplot(x='player_positions', y='avg_player_ratings', data=grouped_position.head(15))
plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
````
