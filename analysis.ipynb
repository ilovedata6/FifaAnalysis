Since the columns in the dataset are a total of 104, it would be overwhelming to perform analysis on all of them at once. Instead, It would be better to combine some features that hold the same meaning. For example, - lwb, ldm, cdm, rdm, rwb, lb, lcb, cb, rcb, rb are 9 defending position columns so, we can combine all these to one column `defensive_positions`. This would reduce the number of columns from 104 to 95.

### columns that need to be reduced : 

- lwb, ldm, cdm, rdm, rwb, lb, lcb, cb, rcb, rb : `defensive_positions`
- ls, st, rs, lw, lf, cf, rf, rw, lam, cam, ram, lm, lcm, cm, rcm, rm : `offensive_positions`
- attacking_crossing, attacking_finishing,attacking_heading_accuracy, attacking_short_passing, attacking_volleys : `attacking_Capabilities`
- skill_dribbling, skill_curve, skill_fk_accuracy, skill_long_passing, skill_ball_control : `skill_Capabilities`
- movement_acceleration, movement_sprint_speed, movement_agility, movement_reactions, movement_balance : `movement_Capabilities`
- power_shot_power, power_jumping, power_stamina, power_strength, power_long_shots : `power_Capabilities`
- mentality_aggression, mentality_interceptions, mentality_positioning, mentality_vision, mentality_penalties, mentality_composure : `mental_Capabilities`
- defending_marking, defending_standing_tackle, defending_sliding_tackle : `defending_Capabilities`
- goalkeeping_diving, goalkeeping_handling, goalkeeping_kicking, goalkeeping_positioning, goalkeeping_reflexes : `goalkeeping_Capabilities`

````python
goalkeeping_Capabilities = (df['gk_diving'] + df['gk_handling'] + df['gk_kicking'] + df['gk_reflexes'] + df['gk_speed'] + df['gk_positioning'])/6
goalkeeping_Capabilities
````

````python
goalkeeping_Capabilities = round(goalkeeping_Capabilities,1)
````

````python
defending_Capabilities = (df['defending_sliding_tackle'] + df['defending_standing_tackle'] + df['defending_marking'])/3
defending_Capabilities
````

````python
defending_Capabilities = round(defending_Capabilities,1)
defending_Capabilities
````

````python
mental_Capabilities = (df['mentality_aggression'] + df['mentality_interceptions'] + df['mentality_positioning'] + df['mentality_vision'] + df['mentality_composure'] + df['mentality_penalties'])/6
mental_Capabilities
````

````python
mental_Capabilities = round(mental_Capabilities,1)
````

````python
attack_Capabilities = (df['attacking_crossing'] + df['attacking_finishing'] + df['attacking_heading_accuracy'] + df['attacking_short_passing'] + df['attacking_volleys'])/5
attack_Capabilities
````

````python
movement_capabilities = (df['movement_acceleration'] + df['movement_sprint_speed'] + df['movement_agility'] + df['movement_reactions'] + df['movement_balance'])/5
movement_capabilities
````

````python
power_capabilities = (df['power_shot_power'] + df['power_jumping'] + df['power_stamina'] + df['power_strength'] + df['power_long_shots'])/5
power_capabilities
````

````python
skill_capabilities = (df['skill_dribbling'] + df['skill_curve'] + df['skill_fk_accuracy'] + df['skill_long_passing'] + df['skill_ball_control'])/5
skill_capabilities
````

````python
def convert_int(x):
    try:
        if "+" in x:
            # Split the string into two parts and convert each part to an integer
            num1, num2 = map(int, x.split("+"))
            # Add the two numbers together and return the result
            return num1 + num2
        else:
            # Convert the string to an integer and return it
            return int(x)
    except (ValueError, TypeError):
        # Ignore the error and return the original string
        return x
````

````python
positions = ['lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb']

for i in positions:
    df[i] = (df[i].apply(convert_int)).fillna(0).astype(int)

defensive_positions = (df['lwb'] + df['ldm'] + df['cdm'] + df['rdm'] + df['rwb'] + df['lb'] + df['lcb'] + df['cb'] + df['rcb'] + df['rb'])/10
defensive_positions
````

````python
positions_attack = ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm']

for i in positions_attack:
    df[i] = (df[i].apply(convert_int)).fillna(0).astype(int)

offensive_positions = (df['ls'] + df['st'] + df['rs'] + df['lw'] + df['lf'] + df['cf'] + df['rf'] + df['rw'] + df['lam'] + df['cam'] + df['ram'] + df['lm'] + df['lcm'] + df['cm'] + df['rcm'] + df['rm'])/16
offensive_positions
````

````python
offensive_positions = round(offensive_positions, 1)
offensive_positions
````

````python
df['defensive_positions'] = defensive_positions
df['offensive_positions'] = offensive_positions
df['attack_Capabilities'] = attack_Capabilities
df['defending_Capabilities'] = defending_Capabilities
df['power_capabilities'] = power_capabilities
df['mental_Capabilities'] = mental_Capabilities
df['goalkeeping_Capabilities'] = goalkeeping_Capabilities
df['movement_capabilities'] = movement_capabilities
df['skill_capabilities'] = skill_capabilities
````

````python
# Displaying Object type value counts of dataset Columns including NANs

def val_count(df):
    
    num_to_display = 10 
    for colname in df.dtypes[df.dtypes == 'object'].index:
        val_counts = df[colname].value_counts(dropna=False) # to see NANs as well
        print(val_counts[:num_to_display])
        if (len(val_counts)>num_to_display):
            print(f' \n only displaying first {num_to_display} of {len(val_counts)} values')
        print('\n\n') # just for more spaces inbetween    
````

#### Q1 - What are the most popular positions among players in the dataset?

#### Q2 - How has the average player rating changed over time?

````python
df['dob'] = pd.to_datetime(df['dob'])
````

````python
df['year'] = df['dob'].apply(lambda x: x.year)
grouped_rating = df.groupby('year')['overall'].mean().reset_index()
grouped_rating.rename(columns={'overall':'average_rating'}, inplace=True)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18,6))
sns.barplot(x='year', y='average_rating', data=grouped_rating)
````

#### Q3 - What are the most common nationalities among players in the dataset?

```python

````python
nationality = df['nationality'].value_counts().head(5)

plt.figure(figsize=(14,8))
explode = (0.1,0,0,0,0)
plt.pie(x=nationality,explode=explode,labels=nationality.index,autopct='%1.1f%%', shadow=True, startangle=300)
plt.show()
````

#### Q4 - What is the distribution of player ratings by position?

```python

````python
grouped_position = df.groupby('player_positions')['overall'].mean().reset_index()
grouped_position.rename(columns={'overall':'avg_player_ratings'}, inplace=True) 
grouped_position.sort_values(by='avg_player_ratings', ascending=False, inplace=True)
grouped_position.head(15)

plt.figure(figsize=(18,6))
sns.barplot(x='player_positions', y='avg_player_ratings', data=grouped_position.head(15))
plt.xlabel('Player Positions')
plt.xticks(rotation=45, horizontalalignment='right')
plt.ylabel('Average Player Ratings')
plt.show()
````

#### Q5 - Is there a relationship between a player's age and their rating?

```python

````python
grouped_age = df.groupby('age')['overall'].mean().reset_index()
grouped_age.rename(columns={'overall':'avg_player_ratings'}, inplace=True)
plt.figure(figsize=(18,6))
sns.lineplot(x='age', y='avg_player_ratings', data=grouped_age)
plt.show()
````

`we can clearly see that there is a negative correlation between age and rating. As the age increases, the rating decreases.`
